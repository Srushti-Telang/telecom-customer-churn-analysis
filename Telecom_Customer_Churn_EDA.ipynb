```python
import numpy as np 
import pandas as pd 

df = pd.read_csv(r"C:\Users\srush\Desktop\Telco IBM Customer Churn Dataset\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head(10)
```

```python
#Check basic info
print(df.shape)
print(df.info)
```

```python
#fix Total charges datatype
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
df
```

```python
#Handle missing values
print(df.isnull().sum()) 
```

```python
df
```

```python
df= df.dropna()
df
```

```python
#Standarized Categorical Values
#Replace no internet servives with No
service_cols = ['OnlineSecurity','OnlineBackup', 'DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
for col in service_cols:
    df[col] = df[col].replace('No internet services','No')

df
```

```python
#Replace 'NO phone services' with 'NO'
df['MultipleLines'] = df['MultipleLines'].replace('No phone service','No')
df
```

```python
#Create new feature (Tenure & Group)
df['Tenure Group'] = pd.cut(df['tenure'],bins=[0,12,36,72],labels=['0-1 Year','1-3 Years','3+ Years'])
df
```

```python
df['MonthlyChargeLevel']=pd.cut(df['MonthlyCharges'],bins=[0,35,70,120],labels=['Low','Medium','High'])
df
```

```python
#Convert Yes/No columns to consistent format
yes_no_cols = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
for col in yes_no_cols:
    df[col]= df[col].replace({'Yes':'Yes', 'No':'No'})
df   
```

```python
#Check duplicates
print("Duplicate rows:",df.duplicated().sum())
df
```

```python
#final checks
print(df.info)
print(df.head())
```

```python
df.columns = df.columns.str.strip().str.replace(' ','_')
df
```
